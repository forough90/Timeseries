
# 📘 Two-Stage Least Squares (2SLS)

---

## 🔹 1. Structural Model

$$
Y = \beta X + u
$$

- $X$ is **endogenous**:
$$
E(Xu) \neq 0
$$

👉 OLS is biased/inconsistent

---

## 🔹 2. Instrument

Introduce $Z$ such that:

### Relevance
$$
E(ZX) \neq 0
$$

### Exogeneity
$$
E(Zu) = 0
$$

---

## 🔹 3. First Stage

Project $X$ on $Z$:

$$
X = \pi Z + v
$$

Estimate:
$$
\hat{X} = P_Z X = Z(Z'Z)^{-1} Z'X
$$

---

## 🔹 4. Second Stage

Replace $X$ with predicted value:

$$
Y = \beta \hat{X} + \varepsilon
$$

---

## 🔹 5. 2SLS Estimator

### Matrix form:
$$
\hat{\beta}_{2SLS} =
(X'P_Z X)^{-1} X'P_Z Y
$$

where:
$$
P_Z = Z(Z'Z)^{-1}Z'
$$

---

### Scalar (single instrument) case:

$$
\hat{\beta}_{IV} =
\frac{E(ZY)}{E(ZX)}
$$

---

## 🔹 6. Why 2SLS Works

Using moment condition:

$$
E(Zu) = 0
$$

Substitute $Y = \beta X + u$:

$$
E(ZY) = \beta E(ZX)
$$

Solve:

$$
\beta = \frac{E(ZY)}{E(ZX)}
$$

---

## 🔹 7. Key Idea

- $Z$ isolates **exogenous variation** in $X$
- First stage extracts that variation:
$$
\hat{X} \propto Z
$$
- Second stage uses only this clean variation

---

## 🔹 8. Interpretation

$$
\beta = \frac{\text{variation in } Y \text{ explained by } Z}
{\text{variation in } X \text{ explained by } Z}
$$

---

## 🔹 9. One-line Insight

👉 2SLS replaces endogenous $X$ with its **exogenous component driven by $Z$**

# 📘 2SLS Interpretation of Structural Shock Identification

---

## 🔹 Step 1: Structural Relationship

We want to estimate:

$$
Y_{i,t+h} = \Theta_{h,i1}\,\varepsilon_{1,t} + \text{other shocks}
$$

But the structural shock $\varepsilon_{1,t}$ is **not observed**.

---

## 🔹 Step 2: First Stage (Instrumental Variable)

We use an instrument $Z_t$:

$$
Y_{1,t} = \pi Z_t + v_t
$$

- $Y_{1,t}$: policy variable (e.g. federal funds rate)  
- $Z_t$: instrument (e.g. policy surprise)

The coefficient satisfies:

$$
\pi \propto E(\varepsilon_{1,t} Z_t)
$$

---

## 🔹 Step 3: Predicted Value

$$
\hat{Y}_{1,t} = \pi Z_t
$$

Thus:

$$
\hat{Y}_{1,t} \propto \varepsilon_{1,t}
$$

👉 The fitted value isolates the **exogenous shock component**

---

## 🔹 Step 4: Second Stage

$$
Y_{i,t+h} = \beta \hat{Y}_{1,t} + u_{t+h}
$$

Substitute:

$$
Y_{i,t+h} = \beta (\pi Z_t) + u_{t+h}
$$

---

## 🔹 Step 5: IV Estimator

From IV theory:

$$
\beta = \frac{E(Y_{i,t+h} Z_t)}{E(Y_{1,t} Z_t)}
$$

Using structural relationships:

$$
E(Y_{i,t+h} Z_t) = \Theta_{h,i1}\,\alpha
$$

$$
E(Y_{1,t} Z_t) = \alpha
$$

So:

$$
\beta = \frac{\Theta_{h,i1}\,\alpha}{\alpha} = \Theta_{h,i1}
$$

---

## 🔹 Automatic Normalization

At impact ($h=0$, $i=1$):

$$
\beta =
\frac{E(Y_{1,t} Z_t)}{E(Y_{1,t} Z_t)} = 1
$$

Thus:

$$
\Theta_{0,11} = 1
$$

---

## 🔹 🔑 Key Insight

- The instrument identifies the shock **up to scale**
- 2SLS divides by $E(Y_{1,t} Z_t)$
- This **automatically normalizes the shock**

👉 No separate normalization is needed

# Genral form

## 🔹 2SLS Representation of LP-IV

We do not observe the structural shock $\varepsilon_{1,t}$, so we use an instrument $Z_t$.

---

### General Structural Model

$$
Y_t = \sum_{h=0}^{\infty} \Theta_h \varepsilon_{t-h}
$$

For variable $i$ at horizon $h$:

$$
Y_{i,t+h} = \Theta_{h,i1}\,\varepsilon_{1,t} + \sum_{j \neq 1} \Theta_{h,ij}\,\varepsilon_{j,t} + \text{lags}
$$

---

### Observational (LP) Equation

$$
Y_{i,t+h} = \beta_h\, Y_{1,t} + \gamma' W_t + u_{t+h}
$$

- $Y_{1,t}$: endogenous policy variable  
- $W_t$: controls (lags of $Y_t$)  
- $u_{t+h}$: composite error term  

---

### Endogeneity Problem

$$
E(Y_{1,t} \, u_{t+h}) \neq 0
$$

because $Y_{1,t}$ contains $\varepsilon_{1,t}$.

---

### Instrumental Variable Assumptions

- Relevance:
$$
E(Z_t \varepsilon_{1,t}) \neq 0
$$

- Exogeneity:
$$
E(Z_t \varepsilon_{j,t}) = 0 \quad \forall j \neq 1
$$

---

### Goal

Identify:
$$
\beta_h = \Theta_{h,i1}
$$

using IV (2SLS).
