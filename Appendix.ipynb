#  Appendix: Core Formulas for LP-IV and SVAR-IV

---

## 1. Structural Representation

### Structural Moving Average (SMA)
$$
    Y_t = \Theta(L)\,\varepsilon_t = \sum_{h=0}^{\infty} \Theta_h \varepsilon_{t-h}
$$

- $Y_t \in \mathbb{R}^n$
- $\varepsilon_t \in \mathbb{R}^m$ (structural shocks)
- $\mathbb{E}[\varepsilon_t \varepsilon_t'] = \Sigma_{\varepsilon\varepsilon}$

---

## 2. Impulse Response Function (IRF)

$$
    \Theta_{h,i1} = \frac{\partial Y_{i,t+h}}{\partial \varepsilon_{1,t}}
$$

Interpretation: Effect of a unit shock $\varepsilon_{1,t}$ on variable $i$ at horizon $h$

---

## 3. Local Projection (LP)

$$
    Y_{i,t+h} = \Theta_{h,i1} \varepsilon_{1,t} + u_{t+h}
$$

- If $\varepsilon_{1,t}$ observed → OLS
- Otherwise → IV needed

---

## 4. LP-IV Regression

$$
    Y_{i,t+h} = \Theta_{h,i1} Y_{1,t} + u_{t+h}
$$

- $Y_{1,t}$ is endogenous
- Use instrument $Z_t$

---

## 5. LP-IV Moment Condition

$$
    \Theta_{h,i1} = \frac{\mathbb{E}[Y_{i,t+h} Z_t]}{\mathbb{E}[Y_{1,t} Z_t]}
$$

---

## 6. LP-IV with Controls

$$
    Y_{i,t+h} = \Theta_{h,i1} Y_{1,t} + \gamma' W_t + u_{t+h}
$$

Residualized version:

$$
    \Theta_{h,i1} = \frac{\mathbb{E}[Y^{\perp}_{i,t+h} Z^{\perp}_t]}{\mathbb{E}[Y^{\perp}_{1,t} Z^{\perp}_t]}
$$

---

## 7. Instrument Validity Conditions (LP-IV)

### Relevance
$$
    \mathbb{E}[\varepsilon_{1,t} Z_t] \neq 0
$$

### Exogeneity
$$
    \mathbb{E}[\varepsilon_{2:n,t} Z_t] = 0
$$

### Lead–Lag Exogeneity
$$
    \mathbb{E}[\varepsilon_{t+j} Z_t] = 0 \quad \forall j \neq 0
$$

---

## 8. VAR Representation

$$
    A(L) Y_t = u_t
$$

$$
    u_t = \Theta_0 \varepsilon_t
$$

---

## 9. Structural MA from VAR

$$
    Y_t = C(L)\Theta_0 \varepsilon_t
$$

where

$$
    C(L) = A(L)^{-1}
$$

---

## 10. SVAR-IV Identification

Impact response:

$$
    \Theta_{0,i1} = \frac{\mathbb{E}[u_{i,t} Z_t]}{\mathbb{E}[u_{1,t} Z_t]}
$$

---

## 11. SVAR-IV IRF

$$
    \Theta_h = C_h \Theta_0
$$

---

## 12. Unit Effect Normalization

$$
    \Theta_{0,11} = 1
$$

Interpretation: a 1-unit shock increases $Y_{1,t}$ by 1 unit

---

## 13. Cumulative Impulse Response

$$
    \Theta^{cum}_{h,i1} = \sum_{k=0}^{h} \Theta_{k,i1}
$$

---

## 14. Forecast Error Variance Decomposition (FEVD)

$$
    FEVD_{h,i1} =
    \frac{\sum_{k=0}^{h-1} \Theta_{k,i1}^2 \sigma^2_{\varepsilon_1}}
    {\mathrm{Var}(Y_{i,t+h} \mid \mathcal{F}_t)}
$$

---

## 15. Historical Decomposition

$$
    \text{Contribution of } \varepsilon_{1,t} \text{ to } Y_{i,t+h} =
    \Theta_{h,i1} \varepsilon_{1,t}
$$

---

## 16. Invertibility Condition

$$
    \varepsilon_t = \text{Proj}(\varepsilon_t \mid Y_t, Y_{t-1}, \dots)
$$

Meaning: shocks can be recovered from observables

---

## 17. Key Insight

- **LP-IV:** robust to non-invertibility but less efficient
- **SVAR-IV:** more efficient but requires invertibility

---


# 📊 Appendix: Statistical Tools

---

## 18. Sample Mean

$$
    \bar{X} = \frac{1}{T} \sum_{t=1}^{T} X_t
$$

---

## 19. Sample Variance

$$
    \mathrm{Var}(X_t) = \frac{1}{T-1} \sum_{t=1}^{T} (X_t - \bar{X})^2
$$

---

## 20. Sample Covariance

$$
    \mathrm{Cov}(X_t, Y_t) = \frac{1}{T-1} \sum_{t=1}^{T} (X_t - \bar{X})(Y_t - \bar{Y})
$$

---

## 21. Autocovariance (lag $k$)

$$
    \gamma_k = \mathrm{Cov}(X_t, X_{t-k}) =
    \frac{1}{T-1} \sum_{t=k+1}^{T} (X_t - \bar{X})(X_{t-k} - \bar{X})
$$

---

## 22. Autocorrelation Function (ACF)

$$
    \rho_k = \frac{\gamma_k}{\gamma_0}
$$

---

## 23. Correlation Coefficient

$$
    \rho_{XY} = \frac{\mathrm{Cov}(X_t, Y_t)}
    {\sqrt{\mathrm{Var}(X_t)\,\mathrm{Var}(Y_t)}}
$$

---

## 24. Linear Projection (Best Linear Predictor)

$$
    \beta = \arg\min_b \mathbb{E}[(Y_t - b X_t)^2]
$$

Solution:

$$
    \beta = \frac{\mathrm{Cov}(X_t, Y_t)}{\mathrm{Var}(X_t)}
$$

---

## 25. OLS Estimator

$$
    \hat{\beta} = \frac{\sum_{t=1}^{T} (X_t - \bar{X})(Y_t - \bar{Y})}
    {\sum_{t=1}^{T} (X_t - \bar{X})^2}
$$

---

## 26. IV Estimator (Single Instrument)

$$
    \hat{\beta}_{IV} =
    \frac{\sum_{t=1}^{T} Z_t Y_t}
    {\sum_{t=1}^{T} Z_t X_t}
$$

Population version:

$$
    \beta = \frac{\mathbb{E}[Z_t Y_t]}{\mathbb{E}[Z_t X_t]}
$$

---

## 27. Covariance-Based Moment Condition

$$
    \mathbb{E}[Z_t u_t] = 0
$$

Equivalent to:

$$
    \mathrm{Cov}(Z_t, u_t) = 0
$$

---

## 28. Serial Correlation

$$
    \mathrm{Cov}(X_t, X_{t-1}) \neq 0
$$

---

## 29. White Noise

$$
    \mathbb{E}[\varepsilon_t] = 0, \quad
    \mathrm{Var}(\varepsilon_t) = \sigma^2, \quad
    \mathrm{Cov}(\varepsilon_t, \varepsilon_{t-j}) = 0 \;\; (j \neq 0)
$$

---

## 30. Stationarity (Weak)

$$
    \mathbb{E}[X_t] = \mu
$$

$$
    \mathrm{Var}(X_t) = \sigma^2
$$

$$
    \mathrm{Cov}(X_t, X_{t-k}) = \gamma_k
$$

---

## 31. Law of Large Numbers (LLN)

$$
    \frac{1}{T} \sum_{t=1}^{T} X_t \xrightarrow{p} \mathbb{E}[X_t]
$$

---

## 32. Central Limit Theorem (CLT)

$$
    \sqrt{T}(\bar{X} - \mu) \xrightarrow{d} \mathcal{N}(0, \sigma^2)
$$

---

## 33. HAC Variance (Newey-West idea)

$$
    \hat{\Omega} =
    \sum_{k=-K}^{K} w_k \hat{\gamma}_k
$$

- Accounts for heteroskedasticity and autocorrelation

---

## 34. Projection Interpretation

$$
    u_t = Y_t - \mathbb{E}[Y_t \mid X_t]
$$

Orthogonality:

$$
    \mathbb{E}[X_t u_t] = 0
$$

---

## 🔑 Key Insight

- **Everything in LP-IV is built on covariance moments**
- Identification relies on:
$$
    \mathrm{Cov}(Z_t, \text{error}) = 0
$$
- Not full independence, only **linear orthogonality**
