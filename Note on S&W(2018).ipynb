### 1- Measurement Error in Structural Shocks

#### Decomposition of the Measured Shock

$$
\tilde{\varepsilon}_t = \varepsilon_t + \nu_t
$$

- $\varepsilon_t$: true structural shock  
- $\nu_t$: measurement error  

---

#### Key Assumption (Orthogonality)

$$
\mathbb{E}[\nu_t \varepsilon_t'] = 0
$$

---

#### Implication for Variance

$$
\mathbb{E}[\tilde{\varepsilon}_t \tilde{\varepsilon}_t']
= \mathbb{E}[(\varepsilon_t + \nu_t)(\varepsilon_t + \nu_t)']
$$

$$
= \mathbb{E}[\varepsilon_t \varepsilon_t']
+ \mathbb{E}[\nu_t \nu_t']
+ \underbrace{\mathbb{E}[\varepsilon_t \nu_t'] + \mathbb{E}[\nu_t \varepsilon_t']}_{=0}
$$

$$
= \Sigma_{\varepsilon\varepsilon} + \Sigma_{\nu\nu}
$$

---

#### Orthogonality with True Shocks

$$
\mathbb{E}[\tilde{\varepsilon}_t \varepsilon_t']
= \mathbb{E}[(\varepsilon_t + \nu_t)\varepsilon_t']
= \mathbb{E}[\varepsilon_t \varepsilon_t']
$$

---

#### Key Condition

$$
\mathbb{E}[\nu_t \varepsilon_t'] = 0
$$

---

#### Interpretation

$$
\nu_t \perp \varepsilon_t
$$

The observed shock is **signal + noise**, where the noise is unrelated to the true shock.

---

#### Key Takeaway

> Measurement error increases variance but does not bias identification as long as it is orthogonal to the true structural shocks.

### 2- Decomposition of Monetary Policy Shock



### True Monetary Policy Shock

$$
\varepsilon_t^{MP}
=
\text{all unexpected policy-related information revealed over time}
$$



### Takeaway

> A measured shock (e.g., high-frequency surprise) captures only a **part** of this total information, not the full structural shock.

$$
Z_t^{HF}
\approx
\underbrace{\varepsilon_t^{ann}}_{\text{announcement surprise}}
+ \nu_t
$$

---

$$
\varepsilon_t^{MP}
=
\underbrace{\text{announcement surprise}}_{\text{Kuttner (2001)}}
+
\underbrace{\text{forward guidance}}_{\text{G\"urkaynak et al. (2005)}}
+
\underbrace{\text{information effects}}_{\text{Jaroci\'nski–Karadi (2020)}}
+
\underbrace{\text{leakage}}_{\text{Lucca–Moench (2015)}}
$$

---

### Interpretation

- **Kuttner (2001)** → surprise in the policy rate at announcement  
- **G\"urkaynak, Sack, Swanson (2005)** → future path / forward guidance  
- **Jaroci\'nski & Karadi (2020)** → central bank information shocks  
- **Lucca & Moench (2015)** → pre-announcement information leakage
